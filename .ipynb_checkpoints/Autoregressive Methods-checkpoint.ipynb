```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
from pandas.plotting import autocorrelation_plot
from sklearn.metrics import mean_squared_error

#Autoregressive

from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
datasetH = pd.read_csv('./data/[1h]BTCUSDT.csv')
datasetM = pd.read_csv('./data/[1m]BTCUSDT.csv')
```

# Autoregressive Moving Average (ARMA)

### datasetH

```python
size = len(XH) - 50
train, test = XH[0:size], XH[size:len(XH)]
history = train.tolist()
predictions = train.tolist()
errors = [0] * len(history)
for t in range(len(test)):
    p, d, q = 1, 0, 1
    model = ARIMA(history, order=(p, d, q))
    model_fit = model.fit()
    prediction = model_fit.forecast()[0]
    predictions.append(prediction)
    true = test[t]
    history.append(true)
    error = np.abs(true - prediction)
    errors.append(error)
    print('predicted={}, expected={}, error={}'.format(prediction, true, error))
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(20, 5), sharex=True)
ax1.plot(history, color='black')
ax1.plot(predictions, color='yellow', label='forecast')
ax1.legend()
ax2.plot(errors, color='red')
plt.title('Error')
plt.show()
```

### datasetM

```python
size = len(XM) - 50
train, test = XM[0:size], XM[size:len(XM)]
history = train.tolist()
predictions = train.tolist()
errors = [0] * len(history)
for t in range(len(test)):
    p, d, q = 1, 0, 1
    model = ARIMA(history, order=(p, d, q))
    model_fit = model.fit() 
    prediction = model_fit.forecast()[0]
    predictions.append(prediction)   
    true = test[t]
    history.append(true)   
    error = np.abs(true - prediction)
    errors.append(error)   
    print('predicted={}, expected={}, error={}'.format(prediction, true, error))
```

# Seasonal Autoregressive Integrated Moving-Average (SARIMA)

### datasetH

```python
size = len(XH) - 50
train, test = XH[0:size], XH[size:len(XH)]
history = train.tolist()
predictions = train.tolist()
errors = [0] * len(history)
for t in range(len(test)):
    p, d, q = 1, 1, 1
    P, D, Q, s = 0, 0, 0, 2
    model = SARIMAX(yH, order=(p, d, q), seasonal_order=(P, D, Q, s))
    model_fit = model.fit()
    prediction = model_fit.forecast()[0]
    predictions.append(prediction)
    true = test[t]
    history.append(true)
    error = np.abs(true - prediction)
    errors.append(error)
    print('predicted={}, expected={}, error={}'.format(prediction, true, error))
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(20, 5), sharex=True)
ax1.plot(history, color='black')
ax1.plot(predictions, color='yellow', label='forecast')
ax1.legend()
ax2.plot(errors, color='red')
plt.title('Error')
plt.show()
```

### datasetM

# Conclusion

### I used the last 100 rows in each database to get visualization of Autoregressive prediction methods. As we can see after some training, The Autoregressive Moving Average (ARMA) method is very similar to The Autoregressive Integrated Moving Average (ARIMA) method, but second method includes also a differencing pre-processing step of the sequence to make the sequence stationary, called integration. They both are quite fast and they are showing a good prediction result. 

### The Seasonal Autoregressive Integrated Moving Average (SARIMA) method is more complicated and slow. It takes mote time and memory to process a prediction becouse it combines the ARIMA model with the ability to perform the same autoregression, differencing, and moving average modeling at the seasonal level.

### SARIMA method is great for univariate time series with trend and/or seasonal components, when ARIMA method is good just for univariate time series with trend and without seasonal components and ARMA method - for univariate time series without trend and seasonal components.
