```python
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
from pandas.plotting import autocorrelation_plot
from sklearn.metrics import mean_squared_error
```

```python
dataseth = pd.read_csv('./data/1hBTCUSDT_part_1.csv')
datasetH = pd.read_csv('./data/2021-1h-BTCUSDT_part_2.csv.xls')
datasetm = pd.read_csv('./data/1mBTCUSDT_part_1.csv')
datasetM = pd.read_csv('./data/2021-1m-BTCUSDT_part_2.csv.xls')
dataseth['Start'] = pd.to_datetime(dataseth["Start"].to_numpy(), unit='ms')
dataseth['Stop'] = pd.to_datetime(dataseth["Stop"].to_numpy(), unit='ms')
datasetH['Start'] = pd.to_datetime(datasetH["Start"].to_numpy(), unit='ms')
datasetH['Stop'] = pd.to_datetime(datasetH["Stop"].to_numpy(), unit='ms')
datasetm['Start'] = pd.to_datetime(datasetm["Start"].to_numpy(), unit='ms')
datasetm['Stop'] = pd.to_datetime(datasetm["Stop"].to_numpy(), unit='ms')
datasetM['Start'] = pd.to_datetime(datasetM["Start"].to_numpy(), unit='ms')
datasetM['Stop'] = pd.to_datetime(datasetM["Stop"].to_numpy(), unit='ms')
```

# datasetH

```python
plt.figure(figsize=(20,5))
plt.plot(dataseth['Stop'], dataseth['Close'], color='black', label='olg data')
plt.plot(datasetH['Stop'], datasetH['Close'], color='red', label='new data')
plt.title('Total chart')
plt.legend()
plt.show()
```

```python
xh = dataseth['Stop'].to_numpy()[-200:]
yh = dataseth['Close'].to_numpy()[-200:]
xH = datasetH['Stop'].to_numpy()
yH = datasetH['Close'].to_numpy()
```

```python
plt.figure(figsize=(20,5))
plt.plot(dataseth['Stop'], dataseth['Close'], color='black', label='old data')
plt.plot(xh, yh, color ='blue', label='data, that we chose as a training basement')
plt.plot(xH, yH, color ='red', label='new data')
plt.title('Total')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(20,5))
plt.plot(xh, yh, color ='blue', label='training fragment')
plt.plot(xH, yH, color='red', label='new data')
plt.title('Experiment')
plt.legend()
plt.show()
```

```python
size = len(yH) - 100
train, test = yh[0:size], yH[size:len(yH)]
history = train.tolist()
predictions = train.tolist()
errors = [0] * len(history)
for t in range(len(test)):
    p, d, q = 1, 1, 1
    model = ARIMA(history, order=(p, d, q))
    model_fit = model.fit()    
    prediction = model_fit.forecast()[0]
    predictions.append(prediction)    
    true = test[t]
    history.append(true)    
    error = np.abs(true - prediction)
    errors.append(error)    
    print('predicted={}, expected={}, error={}'.format(prediction, true, error))
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(20, 5), sharex=True)
ax1.plot(history, color='black')
ax1.plot(predictions, color='yellow', label='forecast')
ax1.legend()
ax2.plot(errors, color='red')
plt.title('Error')
plt.show()
```
