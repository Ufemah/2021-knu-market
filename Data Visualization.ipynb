```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
ds = pd.read_csv("data/[1h]BTCUSDT.csv")
ds
```

```python
plt.figure(figsize=(15,10))
plt.plot(ds['Start'], ds['Open'], c='red')
plt.plot(ds['Stop'], ds['Close'], c='black')
plt.show()
```

```python
plt.figure(figsize=(15, 10))
plt.plot(ds['Start'], ds['Low'], color = 'red')
plt.plot(ds['Start'], ds['High'], color = 'black')
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.plot(ds['Start'], ds['Volume'], color = 'red')
```

```python
plt.hist(ds['Volume'], color='blue')
```

```python
plt.hist(ds['Volume'],bins=100)
```
